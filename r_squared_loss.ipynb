## $R^{2}$ Implementation
If your datasets have different scales (or units), it is better using an error calculation that does not care about the data units. In this notebook, we are showing how to implement: $$R^{2}=(SSR(mean)-SSR(predictions))/SSR(mean)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Constitute Dataset

```python
samples = 1000

data_points_1 = np.arange(0, 1000, 1) + np.random.randn(1000)*100 # Large noise added
data_points_2 = np.arange(0, 1000, 1) + np.random.randn(1000) # Smaller noise added
y = np.arange(0, samples, 1)

plt.figure(figsize = (8,5))
plt.subplot(1, 2, 1)
plt.scatter(y, data_points_1, color='b')
plt.subplot(1, 2, 2)
plt.scatter(y, data_points_2, color='r')
plt.show()
```

## Build a Linear regression Model with constant parameters

```python
w = 1.1
bias = 1.2
predictions = w*y + bias
```

```python
predictions_1 = predictions
predictions_2 = predictions
plt.figure(figsize = (8,5))
plt.subplot(1, 2, 1)
plt.scatter(y, predictions_1, color='b')
plt.subplot(1, 2, 2)
plt.scatter(y, predictions_2, color='r')
plt.show()
```

### Calculate MSE

```python
mse_1 = np.sum((data_points_1-predictions_1)**2)/data_points_1.shape[0]
mse_2 = np.sum((data_points_2-predictions_2)**2)/data_points_2.shape[0]
print(f"MSE values computed are:\nFirst:{mse_1}\nSecond:{mse_2}\nDifference is {np.abs(mse_1-mse_2)}")
```

One expects seeing the almost similar errors, as we use use similar distributions.

```python
SSR_mean_1 = np.sum((data_points_1 - np.mean(data_points_1))**2)
SSR_mean_2 = np.sum((data_points_2 - np.mean(data_points_2))**2)

SSR_pred_1 = np.sum((data_points_1-predictions_1)**2)
SSR_pred_2 = np.sum((data_points_2-predictions_2)**2)

R_squared_1 = (SSR_mean_1-SSR_pred_1)/SSR_mean_1
R_squared_2 = (SSR_mean_2-SSR_pred_2)/SSR_mean_2

print(f"R2 values computed are:\nFirst:{R_squared_1}\nSecond:{R_squared_2}")
```

Model works better for the second dataset!
